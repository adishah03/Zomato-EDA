```python
dfc=pd.read_excel('Country-Code.xlsx')
```

```python
dfc.head()
```

```python
finaldf=pd.merge(df,dfc,on='Country Code',how='left')
```

```python
finaldf.head()
```

```python
cval=finaldf.Country.value_counts().values
```

```python
 cval
```

```python
cname=finaldf.Country.value_counts().index
```

```python
sns.barplot(x=cname,y=cval,data=finaldf)
```

```python
plt.pie(cval[:3],labels=cname[:3],autopct='%1.2f%%')
```

```python
ratings=finaldf.groupby(['Aggregate rating',"Rating color",'Rating text']).size().reset_index().rename(columns={0:'Rating counts'})
```

```python
ratings
```

```python
ratings.sort_values(by=['Rating counts'])
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating counts",data=ratings)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating counts",data=ratings,hue="Rating color")
```
